```python
import pandas as pd
from pathlib import Path
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, LeaveOneOut, KFold
from sklearn.metrics import precision_score, recall_score, accuracy_score, roc_auc_score, roc_curve, precision_recall_curve
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
main_path ='/Users/zubrikhina/Documents/datasets/ЭЭГ агрессия/eeg_extracted-features/'
```

```python
def train_model(path):
    print(path.stem)
    coh_data =pd.read_csv(path)
    le = LabelEncoder()
    y =pd.Series(le.fit_transform(coh_data['target']))
    del coh_data['target']
    files =coh_data['fn']
    del coh_data['fn']

    cv = LeaveOneOut()
    
    #cv = KFold(n_splits=10, shuffle=True, random_state=90)
    X = coh_data.fillna(0)
    y_pred = np.zeros(shape=y.shape)
    y_pred_binary = np.zeros(shape=y.shape)
    for train_idx, test_idx in cv.split(X, y):
        model = LinearSVC(C=0.1)
        model.fit(X.iloc[train_idx, :], y.iloc[train_idx])
        y_pred[test_idx] = model.decision_function(X.iloc[test_idx, :])
        y_pred_binary[test_idx] = model.predict(X.iloc[test_idx, :])
        #y_pred[test_idx] = model.predict(X.iloc[test_idx, :])
    print("ROC AUC: ", roc_auc_score(y, y_pred))
    print("Accuracy: ", accuracy_score(y,y_pred_binary))
    print("--------------------")
    return X
```

```python
all_data =[]
for path in Path(main_path).iterdir():
    if path.stem!=".DS_Store":
        X=train_model(path)
        all_data.append(X)
```
