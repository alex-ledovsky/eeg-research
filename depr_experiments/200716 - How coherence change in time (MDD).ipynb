```python
# HC
df1 = pd.read_csv(data_path + '/Garah rest 17-07.csv', index_col='time')[['f3', 'f4', 't6', 'c3']]
df2 = pd.read_csv(data_path + '/Samedova rest 27-05.csv', index_col='time')[['f3', 'f4', 't6', 'c3']]
df3 = pd.read_csv(data_path + '/Obuhov rest 27-12.csv', index_col='time')[['f3', 'f4', 't6', 'c3']]

# Depression
df4 = pd.read_csv(data_path + '/Anosova rest 18-03.csv', index_col='time')[['f3', 'f4', 't6', 'c3']]
df5 = pd.read_csv(data_path + '/Chernyaev rest 20-05.csv', index_col='time')[['f3', 'f4', 't6', 'c3']]
df6 = pd.read_csv(data_path + '/Merkushev rest 29-06.csv', index_col='time')[['f3', 'f4', 't6', 'c3']]
```

```python
def get_coh_alpha_windowed(df):
    get_coh_alpha = pipeline.features.get_func_by_method('coh-alpha')

    rows = []
    for window_idx in range(50):
        df_window = df.iloc[window_idx*125:(window_idx+10)*125]
        d = get_coh_alpha(df_window)
        d['window_idx'] = window_idx
        rows.append(d)
    res_win_10 = pd.DataFrame(rows)

    rows = []
    for window_idx in range(40):
        df_window = df.iloc[window_idx*125:(window_idx+20)*125]
        d = get_coh_alpha(df_window)
        d['window_idx'] = window_idx
        rows.append(d)
    res_win_20 = pd.DataFrame(rows)

    rows = []
    for window_idx in range(30):
        df_window = df.iloc[window_idx*125:(window_idx+30)*125]
        d = get_coh_alpha(df_window)
        d['window_idx'] = window_idx
        rows.append(d)
    res_win_30 = pd.DataFrame(rows)
    return res_win_10, res_win_20, res_win_30
```

## 3 HC subjects

```python
%matplotlib notebook
```

```python
res_win_10, res_win_20, res_win_30 = get_coh_alpha_windowed(df1)
```

```python
plt.figure(figsize=[10, 5])
plt.title('F3-F4 Coh Alpha Connectivity')
plt.plot(res_win_10['window_idx'], res_win_10['coh_alpha_f3_f4'], label='10sec window')
plt.plot(res_win_20['window_idx'], res_win_20['coh_alpha_f3_f4'], label='20sec window')
plt.plot(res_win_30['window_idx'], res_win_30['coh_alpha_f3_f4'], label='30sec window')
plt.legend()
```

```python
plt.figure(figsize=[10, 5])
plt.title('T6-C3 Coh Alpha Connectivity')
plt.plot(res_win_10['window_idx'], res_win_10['coh_alpha_t6_c3'], label='10sec window')
plt.plot(res_win_20['window_idx'], res_win_20['coh_alpha_t6_c3'], label='20sec window')
plt.plot(res_win_30['window_idx'], res_win_30['coh_alpha_t6_c3'], label='30sec window')
plt.legend()
```
